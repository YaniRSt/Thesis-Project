```python
import os
import re
import pandas as pd
```

```python
repo_path = "."

path_pattern = re.compile(r"A(\d+)\.csv$")

for file in os.listdir(repo_path):
    match = path_pattern.match(file)
    if match:
        idx = int(match.group(1))  # extract index (0–4)
        file_path = os.path.join(repo_path, file)
        print(f"Loaded file: {file} with index: {idx}")

        df = pd.read_csv(file_path)
        round_one_rows = []
        round_two_rows = []
        for row in df.itertuples(index=False):
            if row[-1] == False:
                round_one_rows.append(row)
            elif row[-1] == True:
                round_two_rows.append(row)
        round_one_df = pd.DataFrame(round_one_rows)
        round_two_df = pd.DataFrame(round_two_rows)
        round_one_df.to_csv(f"A{idx}-Round1.csv")
        round_two_df.to_csv(f"A{idx}-Round2.csv")
```

```python
repo_path = "."
path_pattern = re.compile(r"A(\d+)\-Round(\d+)\.csv$")

for file in os.listdir(repo_path):
    match = path_pattern.match(file)
    if match:
        idx = int(match.group(1))  # extract index (0–4)
        file_path = os.path.join(repo_path, file)
        print(f"Loaded file: {file} with index: {idx}")
        file_name = file.split(".")[0]
        df = pd.read_csv(file_path)
        df_new = df[["sample_id", "Body"]]
        df_new.to_csv(f"{file_name}-annotator_version.csv")
        print(f"Saved {file_name}-annotator_version.csv")
```
